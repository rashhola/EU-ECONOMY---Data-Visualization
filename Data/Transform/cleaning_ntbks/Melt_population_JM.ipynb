```python
import pandas as pd
```

```python
pop_df = pd.read_csv("population_new.csv")
```

```python
pop_df.head()
```

```python
new_df = pd.melt(pop_df,id_vars=["Country Name", "Country Code"],var_name = 'Year', value_name = 'population')
```

```python
new_df_sort = new_df.sort_values("Country Name")
```

```python
new_df_sort.to_csv("population_melted.csv", index=False)
```
