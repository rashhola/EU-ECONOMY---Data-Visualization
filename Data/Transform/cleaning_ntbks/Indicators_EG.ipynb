```python
import pandas as pd
import pathlib

```

```python
# Store filepath in a variable
GDP_path = pathlib.Path("../Extraction/GDP_per_capita.csv")
```

```python
# Read and display the CSV with Pandas
GDP_file_df = pd.read_csv(GDP_path, encoding="ISO-8859-1")
GDP_file_df
```

```python
# Delete columns: 1960 through 2004, Indicator Name, and Indicator Code)

GDP_DropC_df = GDP_file_df.drop(GDP_file_df.iloc[:, 2:49], axis = 1)
GDP_DropC_df.head()
```

```python
# Delete column (2020)

GDP_Drop2020_df =  GDP_DropC_df.drop(columns="2020")
GDP_Drop2020_df.head()
```

```python
# List all the columns in the table
GDP_Drop2020_df.columns
```

```python
# Load in file
EU_Country_path = "../Extraction/EU_Countries.csv"
```

```python
# Read and display the CSV with Pandas
EU_file_df = pd.read_csv(EU_Country_path)
GDP_file_df.head()
```

```python
# Merge two dataframes using a left join
merge_GDP_df = pd.merge(
    left=EU_file_df, 
    right=GDP_Drop2020_df, 
    on="Country Code", 
    how="left"
)

merge_GDP_df
```

```python
# List all the columns in the table
merge_GDP_df.columns
```

```python
# Delete column (Country Name_y)

merge_GDP_df_Clean =  (merge_GDP_df
.drop(columns="Country Name_y")
.round(decimals=0))
merge_GDP_df_Clean.head()
```

```python
# Rename column (Country Name_x)
Final_GDP_df_Clean = merge_GDP_df_Clean.rename(columns={"Country Name_x":"Country Name"})
Final_GDP_df_Clean
```

```python
# Unpivot three dataframes.

FV_GDP_df = pd.melt(Final_GDP_df_Clean,id_vars=["Country Name", "Country Code"],var_name = 'Year', value_name = 'GDP')
FV_GDP_df
```

```python
FV_GDP_Sort_df = FV_GDP_df.sort_values(['Country Name', 'Year' ], ascending=[True, True])
FV_GDP_Sort_df
```

```python
GDP_lag_df = FV_GDP_Sort_df.groupby(['Country Name'])['GDP'].shift(1)
GDP_lag_df
```

```python
FV_GDP_Sort_df['lastValue'] = ""
FV_GDP_Sort_df
```

```python
FV_GDP_Sort_df['lastValue'] = GDP_lag_df
FV_GDP_Sort_df
```

```python
FV_GDP_Sort_df.loc[FV_GDP_Sort_df['lastValue'].isnull(),'lastValue'] = FV_GDP_Sort_df['GDP']
FV_GDP_Sort_df
```

```python
FV_GDP_Sort_df.to_csv("Resources/Clean_GDP.csv", index=False)
```

```python
# Dependencies
import pandas as pd
import pathlib
```

```python
# Load in file
CPI_path = "../Extraction/CPI.csv"
```

```python
# Read and display the CSV with Pandas
CPI_file_df = pd.read_csv(CPI_path)
CPI_file_df.head()
```

```python
# Delete columns: 1960 through 2004, Indicator Name, and Indicator Code)

CPI_DropC_df = CPI_file_df.drop(CPI_file_df.iloc[:, 2:49], axis = 1)
CPI_DropC_df.head()
```

```python
# Delete column (2020)

CPI_Drop2020_df =  CPI_DropC_df.drop(columns="2020")
CPI_Drop2020_df.head()
```

```python
# List all the columns in the table
CPI_Drop2020_df.columns
```

```python
# Merge two dataframes using a left join
merge_CPI_df = pd.merge(
    left=EU_file_df, 
    right=CPI_Drop2020_df, 
    on="Country Code", 
    how="left"
)

merge_CPI_df
```

```python
# Delete column (Country Name_y)

merge_CPI_df_Clean =  merge_CPI_df.drop(columns="Country Name_y")
merge_CPI_df_Clean.head()
```

```python
# Rename column (Country Name_x)
Final_CPI_df_Clean = merge_CPI_df_Clean.rename(columns={"Country Name_x":"Country Name"})
Final_CPI_df_Clean
```

```python
Final_CPI_df_ND = Final_CPI_df_Clean.round(decimals=2)
Final_CPI_df_ND
```

```python

Final_CPI_pc_df = (Final_CPI_df_ND
 .set_index(["Country Name", "Country Code"])
 .pct_change(axis='columns')
 .drop(['2005'], axis = 1))
Final_CPI_pc_df
```

```python
Final_CPI_pc_NI_df = (Final_CPI_pc_df
.reset_index()
.round(decimals=3))
Final_CPI_pc_NI_df
```

```python
Final_CPI_pc_NI_df[Final_CPI_pc_NI_df.select_dtypes(include=['number']).columns] *= 100
Final_CPI_pc_NI_df
```

```python
# Unpivot three dataframes.

FV_CPI_df = pd.melt(Final_CPI_pc_NI_df,id_vars=["Country Name", "Country Code"],var_name = 'Year', value_name = 'CPI')
FV_CPI_df
```

```python
FV_CPI_Sort_df = FV_CPI_df.sort_values(['Country Name', 'Year' ], ascending=[True, True])
FV_CPI_Sort_df
```

```python
FV_CPI_Sort_df.to_csv("Resources/Clean_CPI.csv", index=False)
```
